train and evaluate a logistic regression classifier


```python
from sklearn.model_selection import train_test_split

# Assuming 'Attrition' is your target variable
y = employee_df['Attrition']

# Now split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print('Accuracy {} %' . format(100 * accuracy_score(y_pred, y_test) ))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import tensorflow as tf

# Create a Sequential model
model = tf.keras.models.Sequential()

# Adjust input_shape to match the actual number of features in x_train
model.add(tf.keras.layers.Dense(units=500, activation='relu', input_shape=(x_train.shape[1],)))
# x_train.shape[1] will give you the correct number of features

model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Compile the model immediately after defining it
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # This line was missing

history = model.fit(x_train, y_train, epochs=100, batch_size=50)# validation_data=(x_test, y_test))
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
plt.plot(history.history['accuracy'])
#plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
